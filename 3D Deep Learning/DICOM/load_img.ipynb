```python
import pydicom as dicom
import numpy as np
import matplotlib.pylab as plt
from pathlib import Path
```

```python
BASE_PATH=Path.cwd()
FILE_PATH=BASE_PATH/"D0001.dcm"
```

```python
img=dicom.dcmread(FILE_PATH)
print(img)
```

```python
print(f"The Patient ID is:{img.PatientID} ")
print(f"The Study Date is:{img.StudyDate} ")
print(f"The Modality is:{img.Modality} ")
print(f"The number of rows in the DICOM image is:{img.Rows} ")
print(f"The number of columns in the DICOM image is:{img.Columns} ")
print(f"The pixel spacing in DICOM image is:{img.PixelSpacing} ")
print(f"The Slice Thickenss in DICOM image is:{img.SliceThickness} ")
print(f"The Instance Number of DICOM image is:{img.InstanceNumber} ")
```

```python
plt.imshow(img.pixel_array,cmap=plt.cm.gray)
plt.tight_layout()
plt.axis('off')
plt.show()
```
