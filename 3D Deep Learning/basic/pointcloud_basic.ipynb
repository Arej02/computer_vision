```python
from pathlib import Path
import open3d as o3d
import numpy as np
```

```python

DATA_ROOT = Path("/content/drive/MyDrive")
SAMPLE_PATH=DATA_ROOT/"sample.xyz"
```

```python
points=np.loadtxt(SAMPLE_PATH,skiprows=1)
```

```python
points.shape # For x object, there are 1505010 points and 6 features (x,y,z,r,g,b) describing it
```

```python
points[0]
```

```python
points[1][3]
```

```python
xyz=points[:,:3]
```

```python
pcd=o3d.geometry.PointCloud()
pcd.points=o3d.utility.Vector3dVector(xyz)
```

```python
o3d.io.write_point_cloud("cloud.ply", pcd) # Save the 3d plot in .ply format
```
