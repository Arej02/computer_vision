```python
import cv2
from ultralytics import YOLO
import matplotlib.pylab as plt
```

```python
model=YOLO("yolov8n.pt")
```

```python
# 1) Load the image
car=cv2.imread("car.webp")
h,w,c=car.shape
print(h,w,c)
# 2) Send it to the YOLO model
results=model(car,verbose=False)
result=results[0]
# 3) Check the bounding box and draw a rectangle
if result.boxes is not None and len(result.boxes) > 0:
    x1,y1,x2,y2=result.boxes.xyxy[0].cpu().int().tolist()

    cv2.rectangle(car,(x1,y1),(x2,y2),(255,0,0),2)
    # 4) Crop and display
    cropped=car[y1:y2,x1:x2]

    fig,axes=plt.subplots(1,2,figsize=(15,6))

    axes[0].imshow(car)
    axes[0].set_title("Original Car")
    axes[0].axis("off")

    axes[1].imshow(cropped)
    axes[1].set_title("Cropped Car")
    axes[1].axis("off")

```
