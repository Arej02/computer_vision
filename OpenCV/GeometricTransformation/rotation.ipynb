```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
dog=cv2.imread("dog.jpg")
h,w,c=dog.shape
```

```python
print(h,w,c)
```

```python
T=cv2.getRotationMatrix2D((w//2,h//2),180,1) # Rotation Matrix
translated_img=cv2.warpAffine(dog,T,(h,w))

fig,axes=plt.subplots(1,2,figsize=(15,5))

axes[0].imshow(dog)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(translated_img)
axes[1].set_title("Translated Image")
axes[1].axis("off")
```
