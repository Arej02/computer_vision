```python
import numpy as np
import matplotlib.pylab as plt
import cv2
```

```python
dog=cv2.imread("dog.jpg")
```

```python
h,w,c=dog.shape
print(h,w,c)
```

```python
T=np.array([[1,0,200],[0,1,200]],dtype=np.float32) # This is the translation matrix
translated_img=cv2.warpAffine(dog,T,(w,h))
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,6))

axes[0].imshow(dog)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(translated_img)
axes[1].set_title("Translated Image")
axes[1].axis("off")
```
