```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
zeros=np.zeros((100,100))
ones=np.ones((100,100))

rimg=cv2.merge((255*ones,zeros,zeros)).astype(np.uint8)
gimg=cv2.merge((zeros,255*ones,zeros)).astype(np.uint8)
bimg=cv2.merge((zeros,zeros,255*ones)).astype(np.uint8)


plt.figure()

plt.subplot(131)
plt.imshow(rimg)

plt.subplot(132)
plt.imshow(gimg)

plt.subplot(133)
plt.imshow(bimg)

plt.show()
```

```python
bgr_car=cv2.imread("red_car.jpg")
rgb_car = cv2.cvtColor(bgr_car, cv2.COLOR_BGR2RGB)  

B,G,R=cv2.split(bgr_car)

zerosR=np.zeros_like(R)
zerosB=np.zeros_like(B)
zerosG=np.zeros_like(G)

r_img=cv2.merge((zerosB,zerosG,R))
b_img=cv2.merge((B,zerosG,zerosR))
g_img=cv2.merge((zerosB,G,zerosR))


plt.figure()

plt.subplot(151)
plt.imshow(rgb_car)
plt.title("Original Car")

plt.subplot(152)
plt.imshow(bgr_car)
plt.title("BGR Car")

plt.subplot(153)
plt.imshow(r_img)
plt.title("Red Car")

plt.subplot(154)
plt.imshow(b_img)
plt.title("Blue Car")

plt.subplot(155)
plt.imshow(g_img)
plt.title("Green Car")

```
