```python
import cv2
import numpy
import matplotlib.pylab as plt
```

```python
car=cv2.imread("red_car.jpg")
rgb_car=cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
```

```python
hsv=cv2.cvtColor(rgb_car,cv2.COLOR_BGR2HSV)

# Mask to detect blue color:
low_range=numpy.array([101,155,84])
high_range=numpy.array([179,255,255])

mask=cv2.inRange(hsv,low_range,high_range)

plt.figure()

plt.subplot(121)
plt.imshow(rgb_car)
plt.title("Car")
plt.axis('off')

plt.subplot(122)
plt.imshow(mask,cmap='gray')
plt.title("HSV masked Car")
plt.axis('off')

plt.title("HSV mask")
```

```python
leaf=cv2.imread("leaf.jpeg")


hsv=cv2.cvtColor(leaf,cv2.COLOR_BGR2HSV)

# Mask to detect green color:
low_range = numpy.array([40, 50, 50])   
high_range = numpy.array([90, 255, 255]) 

mask=cv2.inRange(hsv,low_range,high_range)

plt.figure()

plt.subplot(121)
plt.imshow(leaf)
plt.title("Leaf")
plt.axis('off')

plt.subplot(122)
plt.imshow(mask,cmap='gray')
plt.title("HSV masked Leaf")
plt.axis('off')

plt.title("HSV mask")
```
